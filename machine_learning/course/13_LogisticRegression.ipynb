```python
import pandas as pd
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import sklearn
```

```python
sklearn.__version__
```

```python
%matplotlib inline
```

```python
dataset = pd.read_csv('MachineLearningAZ/Part3_Classification/Section14_LogisticRegression/Python/Social_Network_Ads.csv')
```

```python
dataset
```

```python
# separate features and columns
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
print(y)
```

```python
# split train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
```

```python
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Use Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=0)
```

```python
classifier = classifier.fit(X_train, y_train)
```

```python
classifier.predict(sc_X.transform([[30, 87000]]))
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), axis=1))
```

```python
# build the confusion matrix, it tells you how many correct and incorrect predictions you had
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# By definition a confusion matrix C is such that C(i,j) is equal to the number of observations known to be in group i and predicted to be in group j
# Thus in binary classification, the count of true negatives is C(0, 0), false negatives is C(1,0), true positives is C(1, 1) and 
# false positives is C(0, 1)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
acc_score = accuracy_score(y_test, y_pred)
acc_score
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
sc = sc_X
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualize test set results
from matplotlib.colors import ListedColormap
sc = sc_X
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
