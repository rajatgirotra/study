```python
# KNN classification --> probably the easiest one
# Step 1: Choose a value for K, Usually K=5
# Step 2: Find the K nearest neighbors from your test data point using Euclidean distance
# Step 3: Count which category each of the K neighbors belong to
# Step 4: Assign the new data point to thte category where you counted the most neighbors.
# Step 5: done
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('MachineLearningAZ/Part3_Classification/Section15_KNN/Python/Social_Network_Ads.csv')
```

```python
dataset
```

```python
%matplotlib inline
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
```

```python
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# import KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
```

```python
classifier = classifier.fit(X_train, y_train)
```

```python
classifier.predict(sc_X.transform([[30, 87000]]))
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), axis=1))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
sc = sc_X
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualize test set results
from matplotlib.colors import ListedColormap
sc = sc_X
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
